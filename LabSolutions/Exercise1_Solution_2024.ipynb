```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

## 1)

```python
import random
DG = nx.DiGraph()
nodes = [i for i in range(1, 51)]
DG.add_nodes_from(nodes)

for node in nodes:
    num_links = random.randint(1, 4)
    for _ in range(num_links):
        #target can't be the same as node 
        temp = nodes.copy()
        temp.remove(node)
        target_node = random.choice(temp)
        DG.add_edge(node, target_node)

    
```

```python
# a)
nx.draw(DG, with_labels=True, font_weight='bold')
```

```python
# b) adjacency matrix

nx.convert.to_dict_of_dicts(DG)
```

```python
# c) Calculate the degree centrality of each node and the average degree of the graph  (use appropriate functions in NetworkX) and display their values

print("Degree centrality of each node: ", nx.degree_centrality(DG))
print("Average degree of the graph: ", np.mean(list(DG.degree()), axis = 0)[1])
```

```python
# d) - Draw the degree distribution plot and comment on whether the power-law distribution is fit 

degree_sequence = sorted([d for n, d in DG.degree()], reverse=True)
degreeCount = {}
for degree in degree_sequence:
    if degree in degreeCount:
        degreeCount[degree] += 1
    else:
        degreeCount[degree] = 1


degree, count = zip(*degreeCount.items())
plt.bar(degree, count, color='b')

# As every node has one to fozr connections the power law does not hold
```

```python
# e) - Test other centrality measures available in NetworkX and display their values, and store the centrality values in a vector

# in degree
print("In degree centrality of each node: ", nx.in_degree_centrality(DG))
in_degree_list = list(nx.in_degree_centrality(DG).values())


# out degree
print("Out degree centrality of each node: ", nx.out_degree_centrality(DG))
out_degree_list = list(nx.out_degree_centrality(DG).values())

# closeness
print("Closeness centrality of each node: ", nx.closeness_centrality(DG))
closeness_list = list(nx.closeness_centrality(DG).values())
```

```python
#  f) - write a script that randomly removes one node from the above graph- Repeat the process a)-d)  until the number of nodes in the graph is equal to one.


x = []
centrality_list = []
in_degree_centrality_list = []
out_degree_centrality_list = []
closeness_centrality_list = []


while len(DG.nodes) > 1:
    node = random.choice(list(DG.nodes))
    DG.remove_node(node)
    print("Degree centrality of each node: ", nx.degree_centrality(DG))
    print("Average degree of the graph: ", np.mean(list(DG.degree()), axis = 0)[1])

    degree_sequence = sorted([d for n, d in DG.degree()], reverse=True)
    degreeCount = {}
    for degree in degree_sequence:
        if degree in degreeCount:
            degreeCount[degree] += 1
        else:
            degreeCount[degree] = 1


    degree, count = zip(*degreeCount.items())
    plt.bar(degree, count, color='b')

    x.append(len(DG.nodes))
    centrality_list.append(np.var(list(nx.in_degree_centrality(DG).values())))
    in_degree_centrality_list.append(np.var(list(nx.in_degree_centrality(DG).values())))
    out_degree_centrality_list.append(np.var(list(nx.out_degree_centrality(DG).values())))
    closeness_centrality_list.append(np.var(list(nx.closeness_centrality(DG).values())))
    #plt.show()
```

```python
#  g ) Display a graph showing the variations of the various centrality measures as a function of the number of edges in the graph.

plt.plot(x, centrality_list, label = "Centrality")
plt.title("Centrality")
plt.show()

plt.plot(x, in_degree_centrality_list, label = "In degree centrality")
plt.title("In degree centrality")
plt.show()

plt.plot(x, out_degree_centrality_list, label = "Out degree centrality")
plt.title("Out degree centrality")
plt.show()

plt.plot(x, closeness_centrality_list, label = "Closeness centrality")
plt.title("Closeness centrality")
plt.show()
```
